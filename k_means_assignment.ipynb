```python
import sys
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
data = pd.read_csv("housing.csv")
data.head()
```

```python
def find_best_kmeans(data, min_k, max_k):

    """  write from here  """
    best_score = float(0)
    best_k = min_k
    # Initialize the K-Means model
    for k in range(min_k, max_k+1):
        km = KMeans(n_clusters=k, n_init='auto', random_state=0)
        km.fit(data)
        labels = km.labels_
        # Use the data and calculate Silhouette Coefficient for the range of K provided
        score = metrics.silhouette_score(data, labels)
        if (score > best_score):
            best_score = score
            best_k = k
    # Return the best K with respect to Silhouette Coefficient
    print("best k:",best_k)
    return best_k
    """        End        """
```

```python
find_best_kmeans(data, 2, 4)
find_best_kmeans(data, 3, 7)
find_best_kmeans(data, 13, 17)
find_best_kmeans(data, 21, 25)
find_best_kmeans(data, 7, 10)
```
